```python
import json

from pathlib import Path
```

```python
dir_path = Path("./VideoLLaMA3-7B/run_1")
out_path = Path("./VideoLLaMA3-7B/out")

out_path.mkdir(exist_ok=True, parents=True)

json_files = list(dir_path.glob("*.json"))
print(json_files)

json_files = {str(name.name).split("_")[3] : str(name) for name in json_files}
print(json_files)
# assert len(json_files) == 22
```

```python
for episode_id, file_path in json_files.items():
    
    json_path = Path(file_path)
    print(f"Cleaning: {json_path.name}")

    # Load JSON data (a list of dicts)
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    filtered_data = [
        entry for entry in data
        if episode_id in entry.get("episode_span", [])
    ]
    
    output_path = out_path / f"{json_path.name}"
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(filtered_data, f, indent=4)
    
    print(f"Filtered file saved to: {output_path}")
```
